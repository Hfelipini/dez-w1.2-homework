```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('green_tripdata_2019-01.csv.gz', nrows=100)
```

```python
df
```

```python
df.lpep_pickup_datetime= pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
print(pd.io.sql.get_schema(df, name='green_trip_data'))
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

```python
print(pd.io.sql.get_schema(df, name='green_trip_data', con=engine))
```

```python
df_iter = pd.read_csv('green_tripdata_2019-01.csv.gz', iterator=True, chunksize=100000)
```

```python
df = next(df_iter)
```

```python
len(df)
```

```python
df.lpep_pickup_datetime= pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

Cria o cabeçalho da tabela

```python
df.head(n=0).to_sql(name='green_trip_data', con=engine, if_exists='replace')
```

```python
%time df.to_sql(name='green_trip_data', con=engine, if_exists='append')
```

```python
from time import time
```

```python
while True:

    t_start = time()
    df = next(df_iter)
    
    df.lpep_pickup_datetime= pd.to_datetime(df.lpep_pickup_datetime)
    df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)

    df.to_sql(name='green_trip_data', con=engine, if_exists='append')

    t_end = time()

    print('inserted another chunk..., took %3.f second' % (t_end - t_start))
```
